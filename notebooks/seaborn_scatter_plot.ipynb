```python
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt

base = pd.read_csv('co2.csv')
base.head()
```

```python
sbn.scatterplot(data=base,x=base.conc,y=base.uptake,hue=base.Type)
```

```python
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
plt.figure()
plt.subplot(1,2,1)
sbn.scatterplot(x=q.conc,y=q.uptake).set_title('Quebec')
plt.subplot(1,2,2)
sbn.scatterplot(x=m.conc,y=m.uptake).set_title('Mississippi')
plt.tight_layout()
```

```python
ch = base.loc[base['Treatment'] == 'chilled']
nh = base.loc[base['Treatment'] == 'nonchilled']

plt.figure()
plt.subplot(1,2,1)
sbn.scatterplot(x=ch.conc,y=ch.uptake).set_title('Chilled')
plt.subplot(1,2,2)
sbn.scatterplot(x=nh.conc,y=nh.uptake).set_title('NonChilled')
plt.tight_layout()
```
