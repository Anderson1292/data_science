```python
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt

base = pd.read_csv('trees.csv')
base.head()
```

```python
sbn.histplot(base.Volume,kde=True,color='blue',stat='density',bins=10).set(title='Volume')
```

```python
dataset = pd.read_csv('chicken.csv')
dataset.head()
```

```python
group = dataset.groupby(['feed'])['weight'].sum()
group
```

```python
#Histograma considerando apenas o 'Casein'
#apenas o grfico de densidade
sbn.kdeplot(dataset.loc[dataset['feed'] == 'casein'].weight).set(title='Casein')
```

```python
#Histograma para o atributo 'horsebean'
sbn.histplot(dataset.loc[dataset['feed'] == 'horsebean'].weight,kde=True,bins=6).set(title='Horsebean')
```

```python
#Histograma para o atributo 'linseed'
sbn.histplot(dataset.loc[dataset['feed'] == 'linseed'].weight,kde=True).set(title='Linseed')
```

```python
#histograma para o atributo 'soybean'
sbn.histplot(dataset.loc[dataset['feed'] == 'soybean'].weight,kde=True,bins=6).set(title='Soybean')
```

```python
#Histograma para o atributo 'sunflower'
sbn.histplot(dataset.loc[dataset['feed'] == 'sunflower'].weight,kde=True).set(title='Sunflower')
```

```python
plt.figure()
plt.subplot(3,2,1)
sbn.histplot(dataset.loc[dataset['feed'] == 'sunflower'].weight,kde=True).set(title='Sunflower')
plt.subplot(3,2,2)
sbn.histplot(dataset.loc[dataset['feed'] == 'soybean'].weight,kde=True,bins=6).set(title='Soybean')
plt.subplot(3,2,3)
sbn.histplot(dataset.loc[dataset['feed'] == 'linseed'].weight,kde=True).set(title='Linseed')
plt.subplot(3,2,4)
sbn.histplot(dataset.loc[dataset['feed'] == 'horsebean'].weight,kde=True).set(title='Horsebean')
plt.subplot(3,2,5)
sbn.histplot(dataset.loc[dataset['feed'] == 'casein'].weight,kde=True).set(title='Casein')
plt.subplot(3,2,6)
sbn.histplot(dataset.loc[dataset['feed'] == 'meatmeal'].weight,kde=True).set(title='Meatmeal')
plt.tight_layout()
```
