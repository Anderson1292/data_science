```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('trees.csv')
dataset.shape
```

```python
dataset.head()
```

```python
graph = np.histogram(dataset.iloc[:,1],bins=6)
graph
```

```python
plt.hist(dataset.iloc[:,1],bins=6)
plt.title('Trees')
plt.xlabel('Frequency')
plt.ylabel('Height')
```

```python
import seaborn as sbn
```

```python
sbn.histplot(dataset.iloc[:,1],kde=False,bins=6,color='blue').set_title('Trees')
```

```python
sbn.kdeplot(dataset.iloc[:,1],color='blue').set_title('Trees') #gráfico de densidade
```

```python
sbn.histplot(dataset.iloc[:,1],kde=True,color='blue',bins=6).set_title("Trees")
```

```python
#Gráficos de dispersão
plt.scatter(x=dataset.Girth,y=dataset.Volume,color='blue',facecolors='none',marker='*')
plt.title('Trees')
plt.xlabel('Circumference')
plt.ylabel('Volume')
```

```python
plt.plot(dataset.Girth,dataset.Volume)
plt.title('Trees')
plt.xlabel('Circumference')
plt.ylabel('Volume')
```

```python
sbn.regplot(x=dataset.Girth,y=dataset.Volume,data=dataset,fit_reg=False,x_jitter=0.3)
```

```python
data = pd.read_csv('co2.csv')
data.head()
```

```python
x = data.conc
y = data.uptake
```

```python
treatment = list(set(data.Treatment))
treatment
```

```python
#Percorre cada tipo de tratamento e cria um gráfico de dispersão
for i in range(len(treatment)):
  id = data.Treatment == treatment[i]
  plt.scatter(x[id],y[id],label=treatment[i])
plt.legend(loc='lower right')
```

```python
#divisão de gráficos
plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(dataset.Girth,dataset.Volume)
plt.subplot(2,2,2)
plt.scatter(dataset.Girth,dataset.Height)
plt.subplot(2,2,3)
plt.scatter(dataset.Height,dataset.Volume,marker='*')
plt.subplot(2,2,4)
plt.hist(dataset.Volume)
```

```python
#BOXPLOT
plt.boxplot(dataset.Volume,vert=False,showfliers=False,notch=True,patch_artist=True)
plt.title('Trees')
plt.xlabel('Volume')
```

```python
#dados por linha
plt.boxplot(dataset)
plt.title('Trees')
plt.xlabel('Data')
```

```python
plt.boxplot(dataset.Volume,vert=False)
plt.boxplot(dataset.Girth,vert=False)
plt.boxplot(dataset.Height,vert=False)
plt.title('Trees')
plt.xlabel('Data')
```
